# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

API documentation:
"categories"
museum (all) = 10027
bar (all) = 13003 
restaurant (all)= 13065

I'm also going to limit the output using param "fields"

```python
for x in range(10, 12):
    bike_ll = bikes["ll"].iloc[x]

    url = "https://api.foursquare.com/v3/places/search"

    params = {
        "ll" : bike_ll,
        "categories" : "13003,10027,13065",
        "radius" : "75",
        "sort":"DISTANCE",
        "limit" : "50",
        "fields" : "name,geocodes,categories,price,distance"
    }

    headers = {
        "Accept": "application/json",
        "Authorization": "fsq3qnvtw345lsOhc+snv7hneg8d02DFLOe2i9PR0u82sp0="
    }

    response10_12 = requests.request("GET", url, params=params, headers=headers)
response10_12
```

```python
jsoned10_12 = response10_12.json()
```

```python
num_responses10_12 = len(jsoned10_12["results"])
num_responses10_12
```

```python
# now extract all_results into columns:

POI_name = []
latitude = []
longitude = []
category_name = []
category_id = []
prices = []
distance = []
bike_ll = []

for j in jsoned10_12["results"]:
    POI_name.append (j["name"] ),
    distance.append (j["distance"]),
    latitude.append (j["geocodes"]["main"] ["latitude"]),
    longitude.append (j["geocodes"]["main"]["longitude"] ),
    category_name.append (j["categories"][0]["name"] ),
    category_id.append (j["categories"][0]["id"] ),
    bike_ll.append (bikes["ll"].iloc[x])
    
    if "price" in j:
        prices.append(j ["price"])
    else:
        prices.append ("NaN")
```

```python
df10_12 = pd.DataFrame (list (zip (POI_name, latitude, longitude, distance, category_name, category_id, prices, bike_ll)),
    columns = ["POI_name", "latitude", "longitude", "distance", "category_name", "category_id", "prices", "bike_ll"]  )
df10_12
```
