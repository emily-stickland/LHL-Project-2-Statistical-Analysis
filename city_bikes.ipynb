```python
cdmx = requests.get("http://api.citybik.es/v2/networks/ecobici").json()
cdmx
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

Searching the uid returns 311 results

```python
stations = cdmx["network"]["stations"]

uid = []
latitude = []
longitude = []
empty_slots = []
free_bikes = []

for s in stations:
    uid.append(s["extra"] ["uid"])
    latitude.append(s["latitude"])
    longitude.append(s["longitude"])
    empty_slots.append(s["empty_slots"])
    free_bikes.append(s["free_bikes"])
```

Put your parsed results into a DataFrame.

```python
df = pd.DataFrame (list (zip (uid, latitude, longitude, empty_slots, free_bikes)),
    columns = ["uid", "latitude", "longitude", "empty_slots", "free_bikes"] )

# can also do this by making a dictionary: df = ( {"ID" : uid, "Latitude" : latitude etc} )

df.head()
```
